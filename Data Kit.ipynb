1. Numpy
- Create an array with shape (4, 3) of: a. all zeros b. ones c. numbers from 0 to 11
- Tabulate the following function: F(x)=2x^2+5, x∈[1,100] with step 1.
- Tabulate the following function: F(x)=e^−x, x∈[−10,10] with step 1.

1.1. Creating arrays:

```python
import numpy as np

# a. Creating an array of all zeros with shape (4, 3)
zeros_array = np.zeros((4, 3))
print("Array of all zeros:")
print(zeros_array)

# b. Creating an array of all ones with shape (4, 3)
ones_array = np.ones((4, 3))
print("Array of all ones:")
print(ones_array)

# c. Creating an array with numbers from 0 to 11 with shape (4, 3)
sequence_array = np.arange(12).reshape((4, 3))
print("Array with numbers from 0 to 11:")
print(sequence_array)
```

1.2. Tabulating the function F(x) = 2x^2 + 5, for x ∈ [1, 100] with step 1:

```python
import numpy as np

x = np.arange(1, 101)  # Create an array from 1 to 100
f_x = 2 * x**2 + 5  # Calculate the values of the function

# Print the table
print("x\tF(x)")
for i in range(len(x)):
    print(f"{x[i]}\t{f_x[i]}")
```

1.3. Tabulating the function F(x) = e^(-x), for x ∈ [-10, 10] with step 1:

```python
import numpy as np

x = np.arange(-10, 11)  # Create an array from -10 to 10
f_x = np.exp(-x)  # Calculate the values of the function

# Print the table
print("x\tF(x)")
for i in range(len(x)):
    print(f"{x[i]}\t{f_x[i]}")
```

All together:

```python
import pandas as pd

# Importing the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
df = pd.read_csv(url)

# Selecting only the "Team," "Yellow Cards," and "Red Cards" columns
selected_columns = ["Team", "Yellow Cards", "Red Cards"]
selected_data = df[selected_columns]

# Determining the number of teams that participated in the Euro2012
num_teams = selected_data["Team"].nunique()

# Filtering teams that scored more than 6 goals
filtered_data = df[df["Goals"] > 6]

# Printing the results
print("Selected columns:")
print(selected_data)

print("\nNumber of teams participated in the Euro2012:", num_teams)

print("\nTeams that scored more than 6 goals:")
print(filtered_data["Team"])
```

3.1. Scatter Plot

```python
import seaborn as sns
df = sns.load_dataset("iris")
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
```

3.2. Line Plot:

```python
import seaborn as sns
df = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", data=df)
```

3.3. Bar Plot:

```python
import seaborn as sns
df = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=df)
```

3.4. Histogram:

```python
import seaborn as sns
df = sns.load_dataset("tips")
sns.histplot(data=df, x="total_bill")
```

3.5. Box Plot:

```python
import seaborn as sns
df = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=df)
```

3.6. Violin Plot:

```python
import seaborn as sns
df = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=df)
```

3.7. Heatmap:

```python
import seaborn as sns
import pandas as pd

# Load the 'flights' dataset
df = sns.load_dataset("flights")

# Pivot the dataset to create a matrix of passengers for each month and year combination
flights = df.pivot_table(values='passengers', index='month', columns='year')

# Create the heatmap
sns.heatmap(data=flights, cmap="YlGnBu", annot=True, fmt="d")

```

3.8. Pair Plot:

```python
import seaborn as sns
df = sns.load_dataset("iris")
sns.pairplot(data=df, hue="species")
```
